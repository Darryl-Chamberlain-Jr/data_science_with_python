# Overview of Hypothesis Testing

Generally speaking, a Hypothesis Test makes some assumptions about unknown **population** value(s) [*null hypothesis*]. We then calculate how likely it would be for some specific **population** value(s) to be true [*p-value*] given our sample data. We also report on how likely it is we get a "true positive" hypothesis result [*power*] and how much of an impact the assumption has on our data [*effect size*]. 

The reason there are so many different formulas for a Hypothesis Test is based on how we calculate the p-value. To calculate the p-value, we need to construct a probability density function for our null hypothesis. There are MANY ways to do this depending on the question we want to answer and the assumption we want to make for our null hypothesis!

Let's look at some examples of creating this probability density function.

## Probability Density Function (PDF) for Test Statistic

### Z Statistic (mean example)

In MATH 412, you covered parametric (assuming normal distribution) tests such as a Normal Distribution (which has the associated Z-Test). We create the Z Statistic by subtracting the sample mean $x$ by the population mean $\mu$ (which scales the distribution mean to 0) and dividing by the population standard deviation $\sigma$ (which scales the distribution standard deviation to 1). 

$$z = \frac{x - \mu}{\sigma}$$

The formula still works for comparing across 2 populations: 

$$ z = \frac{(x_1 - x_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}} $$

This makes the test statistic (called a Z statistic) follow the standard normal distribution ($\mu=0$ and $\sigma^2=1$).


![Normal Distribution](images/standard_normal_distribution.png)

### T Statistic (mean example)

If we know the population is normally distributed but don't know the population standard deviation, we cannot scale the test statistic to follow the standard normal distribution. Instead, we approximate the population standard deviation with the sample standard deviation divided by the sample error ($\sigma^2 \approx \frac{s^2}{n}$) and scale with that. 

$$ t = \frac{x - \mu}{\frac{s}{\sqrt{n}}} $$

The formula still works for comparing across 2 populations:

$$ \sigma_1 \neq \sigma_2: \,\, t = \frac{(x_1 -x_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{s_1^2}{n_1} - \frac{s_2^2}{n_2}}} $$

$$ \sigma_1 = \sigma_2: \, \, t = \frac{(x_1 -x_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{s_p^2}{n_1} + \frac{s_p^2}{n_2}}} $$

where $s_p$ is the pooled standard deviation. 

This makes the test statistic (called a T statistic) follow a T distribution with degrees of freedom $\nu = n-1$. In the picture below we can see this will get closer to the standard normal distribution as our sample size increases. 


![T Distribution](images/student_t_distribution.png)

### Chi-Squared Statistic

If we don't want to make an assumption about the population distribution, we can use a nonparametric test that creates a known distribution for the test statistic. In Math 412, you learned about Chi-Squared Test that bins values into classes to consider the propability an observation falls into a bin. These probabilities are now random variables, so we can calculate the Chi-Squared Test Statistic with $k$ degrees of freedom by summing the square of these random variables, which follows a Chi-Squared Distribution. 

$$ \chi^2 = \sum_{i=1}^n \frac{(O_i - E_i)^2}{E_i} $$

$O_i = \text{ the number of observations of type } i$

$E_i = \text{ the number of expected frequencies of type } i \text{ based on the assumption of the null hypothesis} $

Unlike the T distribution, we can see the degrees of freedom significantly changes the shape of the distribution!


![Chi-Squared Distribution](images/chi_squared_dist.png)

## Integrate PDF for p-value

Once we create the Probability Density Function (PDF) for our test statistic, we can integrate to find the probability of the test statistic occuring under the null hypothesis assumption. The alternative hypothesis tells us how to integrate (e.g., are we integrating the left tail, right tail, or both tails). 

## Reject or Fail to Reject and Power

We either:
- **Reject the Null Hypothesis** as we have statistically significant evidence the null hypothesis assumption is not true OR
- **Fail to Reject the Null Hypothesis** as we do not have statistically significant evidence the null hypothesis assumption is not true

|                  |   **Reject**     |  **Fail to Reject** |
|------------------|------------------|---------------------|
|**Null Hyp False**| True Positive    | Type 2 Error        |
|**Null Hyp True** | Type 1 Error     | True Negative       |

**Power** is the probability of a rejecting a false null hypothesis. A common goal for power is at least 0.8 and ideally at least 0.9. 

It frequently helps students to think about this like a court room verdict: 
- **Guilty** as we have evidence beyond a shadow of a doubt that the null hypothesis assumption (not guilty) is not true
- **Not Guilty** as we do not have evidence beyond a shadow of a doubt that the null hypothesis assumption (not guilty) is not true

|                  |   **Guilty**     |   **Not Guilty**    |
|------------------|------------------|---------------------|
|**Did It**        | True Conviction  | False Acquittal     |
|**Did Not Do It** | False Conviction | True Acquittal      |

In this scenario, **Power** is the probability of a true conviction. Keeping the same goals as before, this would mean at minimum an 80% chance of a true conviction and ideally at minimum a 90% chance of a true conviction. 

## Effect Size

The court room only has 2 options: the person did the crime or they did not. Our null hypotheses are not commonly a binary result though. So how do we determine how much of a difference we can detect when we talk about the null hypothesis being false? This is where **Effect Size** comes into play. 

For example, consider the null hypothesis $\mu = 9.7$ and alternative hypothesis $\mu < 9.7$. Let's say we reject the null hypothesis. A reasonable question is then "how much below mean of 9.7 is our result?". Effect Size is calculated using Power, Sample Size, and Significance Level. This lets us use this calculation in two ways:

- **A Priori** (before conducting a study): By setting acceptable levels of Power, Effect Size, and Significance Level, we can calculate a minimum Sample Size to collect. 

- **Post Hoc** (after conducting a study): By setting an acceptable level of Power and Significance Level, we can use the Sample Size to calculate the Effect Size we can detect. 

Common formulas for Effect Size are:

Cohen's $d$ (Z/T Tests): 

$$(1 \text{ Pop}): \, \, d = \frac{\bar{x} - \mu}{\sigma}$$

$$(2 \text{ Pop}): \, \, d = \frac{\bar{x_1} - \bar{x_2}}{\sigma}$$

For T-Tests, we need to estimate $\sigma$.

Cramer's $V$ (Chi-Squared Contingency Table)

$$ V = \sqrt{\frac{\chi^2}{n \cdot df}} $$

We can use the following categories for Effect Sizes:

| Effect Size | $d$ | $V$ |
|-------------|-----|-----|
|    Small    | 0.2 | 0.1 |
|    Medium   | 0.5 | 0.3 |
|    Large    | 0.8 | 0.5 |


Going back to our example of $\mu = 9.7$ and alternative hypothesis $\mu < 9.7$, if we reject the null hypothesis with small effect size, it tells us that we did have evidence $\mu < 9.7$ and it is likely only a little under 9.7.

This can be extremely useful with Chi-Squared tests as it tells us how much a difference there is between two or more categories.
