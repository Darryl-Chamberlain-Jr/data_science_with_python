# General Steps for Hypothesis Test

## Step 1: Determine the Null and Alternative Hypothesis

### Null Hypothesis - Assumption we are testing

**Examples**
Z/T Tests:
- The population mean equals some value *($\mu = 6.7$)*
- Means from two different populations are equal *($\mu_1 = \mu_2$)*
- The population proportion equals some value *($p = 0.7$)*
- Population proportions from two different populations are equal *($p_1 = p_2$)*

Example Chi-Squared Tests:
- Proportions from categories A and B are the same *(An equal number of people do and do not like pineapple on pizza)*
- There is no relationship between variables A and B *(Watching college sports as child did not influence how students' decided which college to attend)*

### Alternative Hypothesis - "Opposite" assumption of our null hypothesis

**Examples**
Z/T Tests:
- The population mean is not equal to some value *($\mu \neq 6.7$)*
- One population mean is larger than the other *($\mu_1 > \mu_2$)*
- The population proportion is less than some value *($p < 0.7$)*
- Population proportions from two different populations are different *($p_1 \neq p_2$)*

Example Chi-Squared Tests:
- Proportions from categories A and B are not the same *(More people like pineapple on pizza than not)*
- There is no relationship between variables A and B *(Watching college sports as child did influence how students' decided which college to attend)*

## Step 3: Make Decision

We either:
- Reject the null hypothesis as we have statistically significant evidence the null hypothesis assumption is not true
- Fail to reject the null hypothesis as we do not have statistically significant evidence the null hypothesis assumption is not true 

## Step 4: Calculate Power and Effect Size



### Power

While we normally assume power=0.8 or power=0.9, it is possible to assume a level of effect size and directly calculate power.

**Excel**
- No built-in commands: Manually calculate $\beta$ using .Dist functions

**Python**
- Z 1-Pop: [statsmodels.stats.power.normal_power_het](https://www.statsmodels.org/dev/generated/statsmodels.stats.power.normal_power_het.html)
- Z 2-Pop: [statsmodels.stats.power.NormalIndPower](https://www.statsmodels.org/dev/generated/statsmodels.stats.power.NormalIndPower.html)
- T 1-Pop: [statsmodels.stats.power.TTestPower.power](https://www.statsmodels.org/dev/generated/statsmodels.stats.power.TTestPower.power.html)
- T 2-Pop: [statsmodels.stats.power.TTestIndPower](https://www.statsmodels.org/stable/generated/statsmodels.stats.power.TTestIndPower.html)
- Chi-Squared: [statsmodels.stats.power.GofChisquarePower](https://www.statsmodels.org/dev/generated/statsmodels.stats.power.GofChisquarePower.html)

### Effect Size

**Excel**
- No built-in commands: Manually calculate Cohen's $d$ or Cramer's $V$

**Python**

*Use .solve_power() for the associated power functions above.* This allows you to solve for any one paramter of the power functions. For example: 
- T 2-Pop: [statsmodels.stats.power.TTestIndPower.solve_power](https://www.statsmodels.org/stable/generated/statsmodels.stats.power.TTestIndPower.solve_power.html)
