# Ordinary Least Squares Summary Table

Let's return to the summary table for an Ordinary Least Squares regression:

## Create random data

```python
import numpy as np
import pandas as pd

def convert_to_1d(array):
    return [val[0] for val in array]

# Random data
nob = 8
X_1_random = np.random.rand(nob,1)
X_2_random = np.random.rand(nob,1)
X_3_random = np.random.rand(nob,1)
secret_random_x = np.random.rand(nob,1)

# Our random y-value is x_1 + x_2 + a secret random value
# So we expect x_1 and x_2 to be statistically significant to the model and x_3 to not be
y_random = X_1_random + X_2_random + secret_random_x

X_values = pd.DataFrame(
    {
        'x1': convert_to_1d(X_1_random), 
        'x2': convert_to_1d(X_2_random), 
        'x3': convert_to_1d(X_3_random)
    })
y_values = convert_to_1d(y_random)
```

## Run an Ordinary Least Squares Regression

```python
import statsmodels.api as sm

X_values_with_constant = sm.add_constant(X_values) # Adds a constant coefficient column to estimate beta_0
regression = sm.OLS(y_values, X_values_with_constant)
fitted_regression = regression.fit()

print(fitted_regression.summary())
```

Focusing on the results, we see a few different things in the right corner:
- R-squared
- Adj. R-squared
- F-statistic
- Prob (F-statistic)

# R-Squared

$R^2$ (also written as R-squared) is the proportion of variance explained by the linear regression model. 

$$ R^2 = \frac{\text{TSS} - \text{RSS}}{\text{TSS}} $$

Where 

TSS $= \sum(y_i-\bar{y})^2$ is the Total Sum of Squares and 

$ \text{RSS } = \sum (y_i - \beta_1 x_i - \beta_0)^2 $ is the Residual Sum of Squares.
<br><br>

Since TSS is the total variance in $Y$ and RSS is the amount of **unexplained** variance in $Y$, (TSS - RSS) is our **explained** variance. We then divide by TSS to scale it between 0 and 1. 

# Adjusted R-Squared

Adj. R-Squared adjusts for the number of predictor variables in the model by dividing each sum of squares by its degrees of freedom. 

$$ \text{adjusted } R^2 = 
    \frac{
        \frac{ \text{TSS} }{ \text{df}_{\text{TSS}} } 
        - 
        \frac{ \text{RSS} }{ \text{df}_{\text{RSS}} }
    }
    {
        \frac{ \text{TSS} }{ \text{df}_{\text{TSS}} }
    } 
$$

**But why?** Normally in a linear regression model, adding an additional predictor can only increase your $R^2$. But ideally and realistically we want the least number of predictors to make a model. The adjusted R-Square works by penalizing the addition of predictors that do not substantially change the model's $R^2$. 

```python
X_values_without_x3 = pd.DataFrame(
    {
        'x1': convert_to_1d(X_1_random), 
        'x2': convert_to_1d(X_2_random)
    })

X_values_no_x3_with_constant = sm.add_constant(X_values_without_x3) # Adds a constant coefficient column to estimate beta_0
better_regression = sm.OLS(y_values, X_values_no_x3_with_constant)
better_fitted_regression = better_regression.fit()

print(better_fitted_regression.summary())
```

While $R^2$ has not changed much, removing the unrelated x3 predictor variable has increased our Adjusted $R^2$!

# F-Statistic

Unlike the t-statistics that are specific to each predictor variable, the F-statistic is a test statistic on the entire linear regression model. 

&emsp; $H_0$: There is no relationship between the predictors $x_1, x_2, \ldots x_p$ and $y$. <br>
&emsp; $H_1$: There is some relationship between at least one of the predictors $x_1, x_2, \ldots x_p$ and $y$.

Like before, this can be translated to a statement about the coefficients for each predictor:

&emsp; $H_0$: $\beta_1 = \beta_2 = \ldots = \beta_p = 0$ <br>
&emsp; $H_1$: $\beta_k \neq 0$ for some $k \in [1, p]$

Calculating the F-statistic is similar to calculating the adjusted $R^2$:

$$ F = \frac{(\text{TSS}-\text{RSS})/p}{\text{RSS}/\text{df}_{RSS}} $$

where $p$ is the number of predictors.
