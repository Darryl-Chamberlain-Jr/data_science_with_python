# Random Forest

**Principle**: Random forests aggregate multiple decision trees to improve accuracy and reduce overfitting. <br>

**Working Principle**: 
1.	**Ensemble of Trees:** It builds multiple decision trees by using random subsets of data and features.
2.	**Voting/Averaging:** Each tree in the forest gives a prediction, and the final prediction is determined by aggregating the outputs (voting or averaging) of all the trees. 

**Fundamental Concepts**:
- **Ensemble Learning:** Random forests employ ensemble learning by combining multiple decision trees. This aggregation helps improve accuracy and mitigate overfitting.
- **Random Subsets:** Each tree in the forest is trained on a random subset of the data and features, introducing diversity among trees.
- **Application:** A collection of decision trees, beneficial for high-dimensional data.
- **Strengths:** Reduces overfitting through ensemble learning, robust against noise.
- **Considerations:** Complexity, longer training times for large datasets.


```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Create a random forest classifier
clf_rf = RandomForestClassifier(n_estimators=100)
clf_rf.fit(X_train, y_train)

# Make predictions
predictions_rf = clf_rf.predict(X_test)

# Evaluate accuracy
accuracy_rf = accuracy_score(y_test, predictions_rf)
print("Random Forest Accuracy:", accuracy_rf)
```

```python
fn=iris.feature_names
cn=iris.target_names
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
plot_tree(clf_rf.estimators_[0],
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('clf_rf_individualtree.png')
```
