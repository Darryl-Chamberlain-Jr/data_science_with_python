```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
```

```python
# Load the wine dataset
wine = load_wine()
wine_df = pd.DataFrame(wine.data)
wine_df
```

```python
# Calculate Eigenvalues to determine suggested number of factors
fa = FactorAnalyzer()
fa.fit(wine_df)
ev, v = fa.get_eigenvalues()
n_factors = len(np.where(ev > 1)[0])
print(f'Eigenvalues: {ev}')
print(f'Suggested number of factors: {n_factors}')
```

```python
# Create scree plot using matplotlib
plt.scatter(range(1,wine_df.shape[1]+1),ev)
plt.plot(range(1,wine_df.shape[1]+1),ev)
plt.title('Scree Plot for Wine Data')
plt.xlabel('Number of Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```
